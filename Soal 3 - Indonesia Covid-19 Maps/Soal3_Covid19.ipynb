```python
import pandas as pd
import requests   
import folium as fo
```

__Getting Dataset__

```python
df = pd.read_html('https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia',header=0)
df = df[3]
df = df[['Province','Confirmed','Recovered','Deaths','Active']]
df = df.drop([0,35,36,37])
df = df.replace('Jakarta[b]','Jakarta')
df
```

__Getting latitude, Longitude and Logo__

```python
url = 'https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json'
data = requests.get(url)
if data.status_code == 200:
    provinsi = data.json()
else:
    print('Request Error')
```

```python
df2 = pd.DataFrame(provinsi)
df2.rename(columns={'Provinsi':'Province'}, inplace=True)
df2
```

<hr></hr>

__Merge 2 datafreames__

```python
covid19 = pd.merge(df,df2, on='Province')
covid19.head()
```

<hr></hr>

__Visualize__

```python
peta = fo.Map(location = [-2.275363,99.4192074], ## Lokasi Indonesia
              zoom_start = 4,
            tiles = 'OpenStreetMap'
             )

for i in range(len(covid19)):
    icon = fo.features.CustomIcon(covid19.iloc[i]['logo'], icon_size=(30,30))
    fo.Marker(
    [covid19.iloc[i]['latitude'], covid19.iloc[i]['longitude']],
    popup = 'Total Kasus : '+ str(covid19.iloc[i]['Confirmed']) + ' Pasien Sembuh : ' \
        + str(covid19.iloc[i]['Recovered']) + ' Kematian : ' + str(covid19.iloc[i]['Deaths']),
    tooltip = covid19.iloc[i]['Province'],
    icon = icon
).add_to(peta)

peta
```

__Save as HTML__

```python
peta.save('peta sebaran Covid-19 per provinsi.html')
```
